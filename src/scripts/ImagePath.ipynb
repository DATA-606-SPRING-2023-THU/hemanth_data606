```python
import pandas as pd
import requests
import numpy as np
```

```python
raw_df = pd.read_csv('links.csv')
# dtype={
#     'movieId': 'i4',
#     'imdbId': 'i4',
#     'tmdbId': 'i4'
# }
url = 'https://api.themoviedb.org/3/movie/'
```

```python
raw_df['tmdbId'] = raw_df['tmdbId'].fillna(-1)
```

```python
raw_df
```

```python
raw_df['tmdbId'] = raw_df['tmdbId'].astype(int)
```

```python
raw_df
```

```python
image_path = [] 
for i in raw_df['tmdbId']:
    URL = 'https://api.themoviedb.org/3/movie/'+ str(i) + '?api_key=48335227247ce97ac8190f82cdf782e3'
    r = requests.get(URL)
    if i!= -1 and r.json().get('poster_path'):
        poster_path = 'https://image.tmdb.org/t/p/w500'+ r.json().get('poster_path')
        image_path.append(poster_path)
    else:
        image_path.append('')
```

```python
len(image_path)
```

```python
raw_df['image_path'] = image_path
```
